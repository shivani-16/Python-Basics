```python
import matplotlib.pyplot as plt
import csv
```

```python
x=[]

y=[]

with open('test.csv','r') as csvfile:
    plots=csv.reader(csvfile)
    for col in plots: 
        x.append(col[0])
        y.append(col[1])
        
plt.plot(x,y,label='file')
plt.xlabel('x_axes')
plt.ylabel('y_axes')
plt.title('test graph')

plt.legend()
plt.show()
```
